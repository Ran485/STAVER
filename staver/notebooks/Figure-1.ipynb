```python
def modify_dataframe(df):
    df['count'] = df.count(axis=1)
    df = df.sort_values(by=['count'], ascending=False)
    df = df.drop(['count'], axis=1)
    return df

df = modify_dataframe(df)
```

```python
def caculate_cv(data):
    cv = data.std(axis = 1) / data.mean(axis = 1)
    return pd.DataFrame(cv, columns = ['Coefficient of variation'])
```

```python
def fot_transform(data = None, fill_NA = 'min_div_10'):
    """fill_NA : str : min_div_10 or e-5"""
    if fill_NA == 'min_div_12':
        fill_val = data.min().min()/10
    elif fill_NA == 'e-5':
        fill_val = 0.00001
    # data.replace(np.nan, fill_val, inplace=True)
    data_fot = data*10**11/data.sum()
    # data_fot = data/data.sum()
    # data.replace(np.nan, fill_val, inplace=True)
    return data_fot
```

## 5、测量两个样本蛋白丰度密度分布高斯距离及其 Mann-Whitney U test

```python
mannwhitneyu_density_test = pd.read_csv("/Volumes/T7_Shield/staver/results/DIA_repeat20/processed_data/mannwhitneyu_density_test.csv")
mannwhitneyu_density_test
```

```python
data_cleaned = mannwhitneyu_density_test.dropna()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu

# 假设我们有两个样本的蛋白丰度数据，分别存储在 sample1 和 sample2 中
# sample1 = [1.2, 2.4, 3.6, 4.8, 5.0, 6.2, 7.4, 8.6]
# sample2 = [2.2, 4.4, 6.6, 8.8, 10.0, 12.2, 14.4, 16.6]

# 使用 Mann-Whitney U test 计算两个样本之间的高斯距离
stat, p_value = mannwhitneyu(data_cleaned["STAVER"], data_cleaned["Raw data"])

print("Mann-Whitney U test 的统计值:", stat)
print("Mann-Whitney U test 的 p 值:", p_value)
```
